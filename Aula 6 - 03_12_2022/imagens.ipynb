####  Plotando um gráfico com matplotlib

```python
#matplotlib.org => Ctrl K => pyplot


import matplotlib.pyplot as plt
%matplotlib inline 
# não é mais necessário em versões novas do jupyter
```

```python

#           x           y
plt.plot([3,4,5], [100,200,300])
plt.title('Título')
plt.xlabel('Ano')
plt.ylabel('Alguma medida')
plt.savefig('grafico_linhas.png')
plt.show()
```

```python
#Plotando imagem local
import matplotlib.pyplot as plt
import matplotlib.image as img

imagem = img.imread('frutas.jpg') #imread só carrega imagem local
#display(imagem)
#type(imagem)
#print(imagem.shape) #1414 linhas, 474 colunas, 3 cores primarias (r, g, b)

plt.imshow(imagem)
```

```python
#plotar uma imagem online
from skimage.io import imread

imagem = imread('https://softgraf.com/img/img2.jpg')
print(imagem.ndim, imagem.shape)
plt.imshow(imagem)
plt.show()

```

```python
#convertendo uma imagem para escala de cinza
from skimage.io import imread
#from skimage.tranfosrm import resize
import matplotlib.cm as cm

colorida = imread('https://softgraf.com/cursodatascience/frutas.jpg')
print(colorida.ndim, colorida.shape)
plt.imshow(colorida)
plt.show()


cinza = imread('https://softgraf.com/cursodatascience/frutas.jpg', as_gray=True)
print(cinza.ndim, cinza.shape)
plt.imshow(cinza, cmap=cm.gray) #cmap = Color Map = mapa de cores
plt.show()
```

```python
#recortando imagem
imagem = img.imread('frutas.jpg')

recortada = imagem[:200, :1250]
plt.imshow(recortada)
plt.show()

recortada2 = imagem[-200:, -1250:]
plt.imshow(recortada2)
plt.show()
```

```python
#Reduz o número de pixels a imagem (baixando a resolução)
import matplotlib.image as img
import matplotlib.pyplot as plt
from skimage.transform import resize


imagem = img.imread('frutas.jpg') #(1414, 2119, 3)
#print(imagem.ndim, imagem.shape)
redim = resize(imagem, (141, 211))

plt.imshow(redim)
plt.savefig('img_reduzida.jpg')
plt.show()


plt.imshow(imagem)
plt.show()
plt.close()
```

```python
#convertendo uma matriz de imagem para uma única linha => 3D para 1D
import matplotlib.image as img


imagem = img.imread('frutas.jpg') #(1414, 2119, 3)
img_row = imagem.flatten()
print('Dim:', imagem.ndim, 'shape: ', imagem.shape)
print('Dim:', img_row.ndim, 'shape: ', img_row.shape)
```

```python
#usando pandas com matplotlib - gráfico de barras
#matplotlib.org -> ctrl K -> pyplot.bar

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('cidade-temperatura.xlsm')
display(df)

#mudar o tamanho da figura
fig = plt.figure(figsize=(10,6))

#parâmetros:   X                     Height      Width   Bottom
#plt.bar(['a', 'b', 'c','d','e'], [5,10,15,10,5],  0.5,      0,      align='edge', color='red')


plt.bar(df['Cidade'], df['Temperatura'], color='b')
plt.xticks(df['Cidade'], rotation='vertical')


#visível
plt.grid(True, color='grey', linewidth='1', linestyle='-.')
plt.show()

#salvar o gráfico
fig.savefig('graf-temperatura.png', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
from skimage.io import imread

#carrega imagem como array
img = imread('https://softgraf.com/img/img3.jpg')

#define o tamanho da figura
fig = plt.figure(figsize=(10,6))

#plota a imagem
plt.imshow(img)
plt.show()
               
#a imagem é um arrar 2D
print(img.ndim)
print(img.shape)

#usando fatiamento para remorver o azul da imagem
img_verde = img[:, :, 0] #blue=0
plt.imshow(img_verde)
plt.show()

#aplica mapa de cores
plt.imshow(img_verde, cmap='hot')
plt.show()

plt.imshow(img_verde, cmap='nipy_spectral')
plt.show()

plt.imshow(img_verde, cmap='gray')
plt.show()
```
